# Example: Convergence of Sample Means

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
samples = np.random.normal(loc=5, scale=2, size=10000)

sample_means = [np.mean(samples[:n]) for n in range(1, len(samples)+1)]

plt.figure(figsize=(8,5))
plt.plot(sample_means, label="Sample mean")
plt.axhline(5, color="red", linestyle="--", label="True mean = 5")
plt.xlabel("Number of samples")
plt.ylabel("Sample mean")
plt.title("Convergence of Sample Mean to True Mean (Normal Distribution)")
plt.legend()
plt.show()
```

```python
samples_exp = np.random.exponential(scale=2, size=10000)
sample_means_exp = [np.mean(samples_exp[:n]) for n in range(1, len(samples_exp)+1)]

plt.figure(figsize=(8,5))
plt.plot(sample_means_exp, label="Sample mean")
plt.axhline(2, color="red", linestyle="--", label="True mean = 2")
plt.xlabel("Number of samples")
plt.ylabel("Sample mean")
plt.title("Convergence of Sample Mean to True Mean (Exponential Distribution)")
plt.legend()
plt.show()
```
